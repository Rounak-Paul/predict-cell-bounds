````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, UpSampling2D, concatenate

# Global
DEBUG = True
CMAP = 'magma'
````

### Function
Normalize image data in the closed interval [0,1]

````python
def normalize_nd(arr):
    return (arr - np.amin(arr))/(np.amax(arr) - np.amin(arr))
````

### Function
Given directory path, this function will return numpy array containing combined data from the images in the directory.

Returned numpy array shape: [no. of images, image width, image height]

````python
def filepath_to_ndarray(path):
    output = []
    file_list = list(os.listdir(path))
    file_list.sort(key=lambda x: int(x.split(".")[0].split("_")[1]))
    if DEBUG: print(file_list)
    for file in file_list:
        image = cv2.imread(f'{path}/{file}',cv2.IMREAD_GRAYSCALE)
        output.append(normalize_nd(np.asarray(image,dtype = np.float64)))
    return np.array(output)
    
````

### U-net (CNN architecture)
As the name suggests this U net consists of an ```Encoder```, a ```Decoder``` and ```skip connections``` to combine low-level and high-level features.

#### Encoder
The network starts with the ```Input``` tensor having the shape of the images.

- conv1 is learning 64 features with a 3x3 kernel. Rectified Linear Unit (ReLU) activation function is used as the non negative image data works well with the non negative nature of this activation function. To further refine the features learned conv1 is again passed through the same structured convolution layer. pool1 downsamples the feature map and captures important information.

- conv2 has 128 features and 3x3 kernel. Same as previous step this also has ReLU and in the end MaxPooling is done to get the important features and a downsamples pool2.

#### Middle
In this section a larger kernel, 6x6 is used to capture more complex features. This layers will learn 512 features. 

#### Decoder
Finally the Decoder part starts with the output of the middle layers, conv3. 
- Up sampling is performed, to increase the special dimensions of the feature map. It essentially magnifies the feature map, making it larger.
- up4 is passed through a convolution layer with 128 features and a 2x2 kernel, to learn the features before performing the skip connections. 
- In merge4 the high resolution features from the encoder is introduced. This is where the skip connection is happening.
- In this combined features convolution layers have been applied.
- similar skip connection is done with the conv1 features also.

The model is returned to be compiled and fit with out Train datasets. 




````python
def unet_2(input_shape):
    inputs = Input(input_shape)

    # Encoder
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    # Middle
    conv3 = Conv2D(512, 6, activation='relu', padding='same')(pool2)
    conv3 = Conv2D(512, 6, activation='relu', padding='same')(conv3)

    # Decoder
    up4 = UpSampling2D(size=(2, 2))(conv3)
    up4 = Conv2D(128, 2, activation='relu', padding='same')(up4)
    merge4 = concatenate([conv2, up4], axis=3)
    conv4 = Conv2D(128, 3, activation='relu', padding='same')(merge4)
    conv4 = Conv2D(128, 3, activation='relu', padding='same')(conv4)

    up5 = UpSampling2D(size=(2, 2))(conv4)
    up5 = Conv2D(64, 2, activation='relu', padding='same')(up5)
    merge5 = concatenate([conv1, up5], axis=3)
    conv5 = Conv2D(64, 3, activation='relu', padding='same')(merge5)
    conv5 = Conv2D(64, 3, activation='relu', padding='same')(conv5)

    # Output layer
    outputs = Conv2D(1, 1, activation='sigmoid')(conv5)

    model = Model(inputs=inputs, outputs=outputs)
    return model
````

### U-net (CNN architecture) with less convolution layers

````python
def unet_1(input_shape):
    inputs = Input(input_shape)

    # Encoder
    conv1 = Conv2D(32, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(32, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    # Middle
    conv2 = Conv2D(64, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(64, 3, activation='relu', padding='same')(conv2)

    # Decoder
    up3 = UpSampling2D(size=(2, 2))(conv2)
    up3 = Conv2D(32, 2, activation='relu', padding='same')(up3)
    merge3 = concatenate([conv1, up3], axis=3)
    conv3 = Conv2D(32, 3, activation='relu', padding='same')(merge3)
    conv3 = Conv2D(32, 3, activation='relu', padding='same')(conv3)

    # Output layer
    outputs = Conv2D(1, 1, activation='sigmoid')(conv3)

    model = Model(inputs=inputs, outputs=outputs)
    return model
````

### Compile and Train model
Data is split 80% for Training and 20% for Test + Validation (50% each out of the remaining 20%)

````python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=40)
X_val, X_test, Y_val, Y_test = train_test_split(X_val, Y_val, test_size=0.5, random_state=40)
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],X_train.shape[2], 1))
Y_train = np.reshape(Y_train, (Y_train.shape[0],Y_train.shape[1],Y_train.shape[2], 1))


# Model creation
input_shape = X_train[0].shape
model = unet_2(input_shape)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model training
model.fit(X_train, Y_train, batch_size=4, epochs=100, validation_data=(X_val, Y_val))

# Model evaluation
loss, accuracy = model.evaluate(X_test, Y_test)
print(f"Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}")
````

````python
# Make predictions
predictions = model.predict(X_test)

plt.imshow(X_test[1],cmap=CMAP);plt.title('Test Image'); plt.show()
plt.imshow(Y_test[1],cmap=CMAP);plt.title('Test Lable'); plt.show()
plt.imshow(predictions[1],cmap=CMAP);plt.title('Prediction'); plt.show()
````
